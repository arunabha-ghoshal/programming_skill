```python
from typing import *

def matrixReshape(mat: List[List[int]], r: int, c: int) -> List[List[int]]:
    m = len(mat)
    n = len(mat[0])

    if r*c == m*n:
        r_final = []
        r_col = []
        for row in mat:
            for col in row:
                if len(r_col)<c:
                    r_col.append(col)
                    #print(r_col)
                if len(r_col) == c:
                    r_final.append(r_col)
                    r_col = []
                    #print(r_final)

        return r_final
    else:
        return mat
```

```python
%%timeit -n 100
print('answer :',matrixReshape(mat = [[1,2],[3,4]], r = 1, c = 4))
```
